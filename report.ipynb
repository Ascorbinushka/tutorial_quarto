```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('datasets/Student_performance_data _.csv', encoding = 'windows-1251')
df1 = df[['StudentID','GradeClass', 'ParentalEducation','GPA', 'Absences']]
df1.head(5)
```

Student ID - Уникальный идентификатор, присвоенный каждому студенту (от 1001 до 3392).

Age - Возраст студентов колеблется от 15 до 18 лет.

Gender - Пол студентов, где 0 представляет мужчину и 1 представляет женщину.

Ethnicity - Этническая принадлежность студентов, закодированная следующим образом:

0: Кавказский
1: Афроамериканец
2: Азиатский
3: Другое
ParentalEducation - Уровень образования родителей, закодированный следующим образом:

0: Нет
1: Средняя школа
2: Некоторые колледж
3: Бакалавр
4: Высшее

StudyTimeWeekly - Еженедельное время обучения в часах, в диапазоне от 0 до 20.

Absences - Количество отсутствий в течение учебного года, в диапазоне от 0 до 30.

Tutoring - Статус обучения, где 0 означает «Нет», а 1 означает «Да».

ParentalSupport - Уровень поддержки родителей, закодированный следующим образом:

0: Нет
1: Низкий
2: Умеренный
3: Высокая
4: Очень высокий
Extracurricular - Участие во внеучебной деятельности, где 0 означает Нет, а 1 означает Да.

Sports - Участие в спорте, где 0 означает Нет, а 1 означает Да.

Music - Участие в музыкальной деятельности, где 0 означает Нет, а 1 означает Да.

Volunteering - участие в волонтерской деятельности, где 0 означает «нет», а 1 означает «да».

GPA - средний балл по шкале от 2,0 до 4,0, на которую влияют привычки обучения, участие родителей и внеклассные занятия.

GradeClass - Классификация оценок учащихся на основе GPA:

0: 'A' (GPA > = 3.5)
1: «B» (3,0 < = GPA < 3,5)
2: 'C' (2,5 < = GPA < 3,0)
3: 'D' (2,0 < = GPA < 2,5)
4: 'F' (GPA < 2.0)

# Распределение оценок (GradeClass)

```python
# Создаем новое окно для графика с определённым размером
plt.figure(figsize=(8, 6))
# Создаем столбчатую диаграмму 
sns.countplot(x='GradeClass', data=df)
# Заголовок
plt.title('Распределение оценок')
# Метка для оси X
plt.xlabel('Оценка')
# Метка для оси Y
plt.ylabel('Количество студентов')
# Устанавливаем метки для делений на оси X, преобразуя числовые значения в буквенные обозначения оценок.
plt.xticks(ticks=[0, 1, 2, 3, 4], labels=['A', 'B', 'C', 'D', 'F'])
# Отображаем график
plt.show()
```

# Средний балл (GPA) по уровню образования родителей (ParentalEducation)

```python
plt.figure(figsize=(10, 6))
sns.boxplot(x='ParentalEducation', y='GPA', data=df)
plt.title('Средний балл (GPA) по уровню образования родителей')
plt.xlabel('Уровень образования родителей')
plt.ylabel('Средний балл (GPA)')
plt.xticks(ticks=[0, 1, 2, 3, 4], labels=['Нет', 'Средняя школа', 'Некоторые колледж', 'Бакалавр', 'Высшее'])
plt.show()
```

# Средний балл (GPA) в зависимости от количества пропусков (Absences)

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(x='Absences', y='GPA', data=df)
plt.title('Средний балл (GPA) и количество пропусков')
plt.xlabel('Количество пропусков')
plt.ylabel('Средний балл (GPA)')
plt.show()
```
